# Seaborn Tutorial

## Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

## Import Data

```python
print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

```python
sns.distplot(crash_df['not_distracted'], kde=False)
```

# Join Plot

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='hex')
```

# KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

```python
sns.pairplot(crash_df)


tips_df = sns.load_dataset('tips')
tips_df
```

# Pair Plots

```python
tips_df = sns.load_dataset('tips')

sns.pairplot(tips_df, hue='sex', palette='Blues')
```

# Rug Plot

```python
sns.rugplot(tips_df['tip'])
```

```python
sns.set_style('white')



plt.figure(figsize=(8,4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('darkgrid')



plt.figure(figsize=(8,4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('whitegrid')



plt.figure(figsize=(8,4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('dark')



plt.figure(figsize=(8,4))

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')



plt.figure(figsize=(8,4))


sns.set_context('paper', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')



plt.figure(figsize=(8,4))


sns.set_context('talk', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')



plt.figure(figsize=(8,4))


sns.set_context('poster', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')



plt.figure(figsize=(8,4))


sns.set_context('paper', font_scale=1.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')



plt.figure(figsize=(8,4))


sns.set_context('paper', font_scale=4.4)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')
```

```python
sns.set_style('ticks')



plt.figure(figsize=(8,4))


sns.set_context('paper', font_scale=1.5)

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

sns.despine(left=False, bottom=True)    
```

# Categorical Plots 

## Bar Plots

```python
sns.barplot(x='sex', y='total_bill', data=tips_df)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.median)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.std)
```

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.var)
```
